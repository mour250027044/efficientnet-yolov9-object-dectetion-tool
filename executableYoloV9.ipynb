```python
import os

import pandas as pd

from ultralytics import YOLO

DATASET = "Ev-Battery-Components-2/data.yaml"
EPOCHS = 100
BATCH = 16
IMGSZ = 640
DEVICE = "0,1"

model = YOLO("yolov9s.pt")
r = model.train(
    data=DATASET,
    epochs=EPOCHS,
    batch=BATCH,
    imgsz=IMGSZ,
    device=DEVICE,
    project="runs/yolov9",
    name="yolov9s",
    exist_ok=True,
)
```
