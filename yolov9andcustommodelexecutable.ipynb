```python
import os, shutil, sys, site, subprocess

print("洫ｹ Cleaning broken pandas installation...")

# 1ｸ鞘Ε Find and remove the pandas folder in site-packages
for path in site.getsitepackages() + [site.getusersitepackages()]:
    pandas_path = os.path.join(path, "pandas")
    if os.path.exists(pandas_path):
        shutil.rmtree(pandas_path, ignore_errors=True)
        print(f"泓托ｸ Removed: {pandas_path}")

# 2ｸ鞘Ε Remove all cached bytecode in site-packages
cache_dir = os.path.join(path, "__pycache__")
if os.path.exists(cache_dir):
    shutil.rmtree(cache_dir, ignore_errors=True)
    print(f"洫ｼ Cleared cache: {cache_dir}")

# 3ｸ鞘Ε Force reinstall pandas cleanly without cache
print("\n筮ｸ Reinstalling pandas cleanly...")
subprocess.run([sys.executable, "-m", "pip", "install", "--no-cache-dir", "--force-reinstall", "pandas==2.2.2"], check=False)

# 4ｸ鞘Ε Verify import
print("\n沐 Verifying import...")
try:
    import pandas as pd
    print("笨 Pandas imported successfully!")
    print("沒 Version:", pd.__version__)
    print("沒ｦ Path:", pd.__file__)
except Exception as e:
    print("笶 Still failing:", e)
```
